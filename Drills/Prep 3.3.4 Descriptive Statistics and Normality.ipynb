```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bernuli = np.random.binomial(1, 0.5,100)
plt.hist(bernuli)
plt.axvline(bernuli.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(bernuli.std(), color = "r", linestyle = 'dashed', linewidth = 2)

plt.show()
```

```python
binomial = np.random.binomial(10, 0.4,100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(binomial.std(), color = "r", linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
exp = np.random.exponential(1, 100)
plt.hist(exp)
plt.axvline(exp.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(exp.std(), color = "r", linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
gamma = np.random.gamma(3, 1, 100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(gamma.std(), color = "r", linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
geo = np.random.geometric(0.3, 100)
plt.hist(geo)
plt.axvline(geo.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(geo.std(), color = "r", linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
pois = np.random.poisson(3.5, 100)
plt.hist(pois)
plt.axvline(pois.mean(), color = "r", linestyle = 'solid', linewidth = 2)
plt.axvline(pois.std(), color = "r", linestyle = 'dashed', linewidth = 2)
plt.show()
```
