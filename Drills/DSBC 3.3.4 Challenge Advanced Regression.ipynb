### Ridge Regression

```python
trainsize = int(df.shape[0] / 2)
df_test = df.iloc[trainsize:, :].copy()
df_train = df.iloc[:trainsize, :].copy()
```

```python
Y_train = df_train['LowCrime']
X_train = df_train[keywords]
```

```python
df_train['Violent_Property'] = df_train['ViolentCrime'] * df_train['PropertyCrime']
df_train['Violent_Population'] = df_train['ViolentCrime'] * df_train['Population']
df_train['Property_Population'] = df_train['Population'] * df_train['PropertyCrime']
df_train['Population_sqrt'] = (df_train['Population']) ** .5
df_train['Population3'] = (df_train['Population']) ** 3
```

```python
keywords2 = ['Population', 'ViolentCrime', 'Murder', 'Rape2', 'Robbery',
       'AggravatedAssault', 'PropertyCrime', 'Burglary', 'LarcenyTheft',
       'MotorVehicleTheft', 'Arson3', 'Population2', 'TotalCrime',
       'PopCrimeRatio', 'Violent_Property', 'Violent_Population',
       'Property_Population', 'Population_sqrt', 'Population3']
```

```python
X_train2 = df_train[keywords2]
```

```python
# Fitting a ridge regression model. Alpha is the regularization
# parameter (usually called lambda). As alpha gets larger, parameter
# shrinkage grows more pronounced. Note that by convention, the
# intercept is not regularized. Since we standardized the data
# earlier, the intercept should be equal to zero and can be dropped.

ridgeregr = LogisticRegression(penalty = 'l2')
ridgeregr.fit(X_train, Y_train)
print(ridgeregr.score(X_train, Y_train))
origparams = ridgeregr.coef_[0]
print(origparams)

ridgeregrBig = LogisticRegression(penalty = 'l2')
ridgeregrBig.fit(X_train2, Y_train)
print(ridgeregrBig.score(X_train2, Y_train))
newparams = ridgeregrBig.coef_[0,0:len(origparams)]
print(newparams)

print('\nParameter Estimates for the same predictors for the small model and large model:')
compare = np.column_stack((origparams, newparams))
prettycompare = np.array2string(
    compare,
    formatter={'float_kind':'{0:.3f}'.format})
print(compare)
```

```python
# Test the simpler model with smaller coefficients.
Y_test = df_test['LowCrime']
X_test = df_test[keywords]
print('\nR-squared simple model:')
print(ridgeregr.score(X_test, Y_test))

# Test the more complex model with larger coefficients.
df_test['Violent_Property'] = df_test['ViolentCrime'] * df_test['PropertyCrime']
df_test['Violent_Population'] = df_test['ViolentCrime'] * df_test['Population']
df_test['Property_Population'] = df_test['Population'] * df_test['PropertyCrime']
df_test['Population_sqrt'] = (df_test['Population']) ** .5
df_test['Population3'] = (df_test['Population']) ** 3

# Re-run the model with the new features.
X_test2 = df_test[keywords2]
print('\nR-squared complex model:')
print(ridgeregrBig.score(X_test2, Y_test))
```

### Lasso Regression

```python
# Fitting a ridge regression model. Alpha is the regularization
# parameter (usually called lambda). As alpha gets larger, parameter
# shrinkage grows more pronounced. Note that by convention, the
# intercept is not regularized. Since we standardized the data
# earlier, the intercept should be equal to zero and can be dropped.

lassoregr = LogisticRegression(penalty = 'l1')
lassoregr.fit(X_train, Y_train)
print(lassoregr.score(X_train, Y_train))
origparams = lassoregr.coef_[0]
print(origparams)

lassoregrBig = LogisticRegression(penalty = 'l1')
lassoregrBig.fit(X_train2, Y_train)
print(lassoregrBig.score(X_train2, Y_train))
newparams = lassoregrBig.coef_[0,0:len(origparams)]
print(newparams)

print('\nParameter Estimates for the same predictors for the small model and large model:')
compare = np.column_stack((origparams, newparams))
prettycompare = np.array2string(
    compare,
    formatter={'float_kind':'{0:.3f}'.format})
print(compare)
```

```python
# Test the simpler model with smaller coefficients.
print('\nR-squared simple model:')
print(lassoregr.score(X_test, Y_test))

# Re-run the model with the new features.
print('\nR-squared complex model:')
print(lassoregrBig.score(X_test2, Y_test))
```
