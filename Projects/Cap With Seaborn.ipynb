```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from matplotlib.gridspec import GridSpec
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
mpl.rcParams['font.size'] = 14.0
```

```python
df = pd.read_csv("C:/Users/LOGIN/Documents/Thinkful/weather.csv",low_memory = False)
```

```python
sns.set(style="white",font_scale=5)

g = sns.lmplot(y='MeanTemperatureF', # Variable 1.
               x='MeanWindSpeedMPH', # Variable 2.
               data=df, # Data
               size = 25,
               col='ZIP',               
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("MeanTemperatureF")
g.set_xlabels("MeanWindSpeedMPH")
#plt.title('Scatterplot: Distance vs Time')
plt.show()
```

```python
sns.set(style="whitegrid",font_scale=2)
plt.figure(figsize=(15, 10))


ax = sns.boxplot(x='ZIP',y='MeanWindSpeedMPH',hue='Events',data=df,palette='pastel')  


plt.title('Relationship between weather and wind speed')
sns.despine(offset=10, trim=True)
ax.set(xlabel='ZIP', ylabel='WindSpeed')
ax.legend(loc='upper left',bbox_to_anchor=(0, 1.3))

plt.show()
```
