# Choose one variable and plot that variable four different ways.

```python
print('Line plots show data over time')
plt.plot(df.iloc[0:10]['price'])
plt.show()

print('Scatterplots show the relationship between two variables.')
plt.scatter(df['price'],df['bedrooms'])
plt.show()

print('Histograms show the distribution of the dataset and any outliers.')
plt.hist(df['price'])
plt.show()

print('Boxplots are used to compare groups and to identify differences in variance, as well as outliers.')
plt.boxplot(df['price'])
plt.show()
```

# Plot two continuous variables, and plot them three different ways

```python
sns.set(style="whitegrid")

g = sns.factorplot(x="waterfront", y="price", hue="condition", data=df,
                   size=6, kind="point", palette="pastel",dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Prices")
g.set_xlabels("With or Without waterfront")
plt.title('Pointplot: Prices by conditons and waterfront')
plt.show()
```

```python
sns.set(style="whitegrid")

ax = sns.boxplot(x='waterfront',y='price',hue='condition',data=df,palette='pastel')  

plt.title('Boxplot: Prices by conditons and waterfront')

sns.despine(offset=10, trim=True)
ax.set(xlabel='With or Without waterfront', ylabel='Prices')

plt.show()
```

```python
sns.set(style="darkgrid")

barplot = sns.factorplot(x="waterfront", y="price", hue="condition", data=df,
                   size=6, kind="bar", palette="pastel")
barplot.despine(left=True)
barplot.set_ylabels("Prices")
barplot.set_xlabels("With or Without waterfront")
plt.title('Barplot: Prices by conditons and waterfron')
plt.show()
```
