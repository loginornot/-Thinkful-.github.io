```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/LOGIN/Documents/Python/top-30-countries-energy-consumption.csv")
df = df.sort_values(['Country'])
df.index = range(320)
df
#Read in the csv file and sort it alphabetically by the country names,then re-indexing the dataframe from 0
```

# Who consumes the most?

```python
df2 = df.groupby(["Country"]).sum()["value in million tons of oil equivalent"] #Group by the countries and their total energy usage from 2007-2016
df2 = pd.DataFrame(df2)        #change to panda dataframe
df2.columns = ["Total Usage"]  #name the colunm
df2 = df2.drop("Global total")   #We don't need Global total here, we just want to see which country consumed most of the energy
df2["Total Usage"].plot(kind='bar',figsize = (16,10))

#As we can see here, China and USA are the top 2 countries that consumed most of the energry during 2007-2016.
#India and The Russian Federation are the two tier-2 countries on energy consumption.
#We have 31 countries here, but there are total 195 countries in the world,so we can see that either China or USA consume more than the rest 164 countries combine.
```

```python
df4 = df.groupby(["Country"]).mean()["value in million tons of oil equivalent"]
df4 = pd.DataFrame(df4)
df4.columns = ["Average Usage"]
df4 = df4.drop("Global total")
df4["Average Usage"].plot(kind='bar',figsize = (16,10))
plt.ylabel("Total energy consumption during 2007-2016")
#This is a histogram that represent the average yearly energy consumption of each country
#Since this is almost the same as the last graph, so we may can come to a conclusion that for each country their yearly energy consumption may stay at the same level
#We can use line plot to verify this
```

# Future Consumption Level Tendency

```python
df5 = df.drop(["YoY in 2016"],axis = 1)              #We don't need this column

cn = df5[(df5['Country'] == 'China')].sort_values(["Year"])     #A dataframe only contain data about China, and sorted according to year
cnx = cn["Year"]                            
cny = cn["value in million tons of oil equivalent"]

us = df5[(df5['Country'] == 'United States')].sort_values(["Year"])   #A dataframe only contain data about USA, and sorted according to year
usx = us["Year"]
usy = us["value in million tons of oil equivalent"]

trf = df5[(df5['Country'] == 'The Russian Federation')].sort_values(["Year"])  #A dataframe only contain data about TRF, and sorted according to year
trfx = trf["Year"]
trfy = trf["value in million tons of oil equivalent"]

ind = df5[(df5['Country'] == 'India')].sort_values(["Year"]) #A dataframe only contain data about India, and sorted according to year
indx = ind["Year"]
indy = ind["value in million tons of oil equivalent"]

jp = df5[(df5['Country'] == 'Japan')].sort_values(["Year"])  #A dataframe only contain data about Japan, and sorted according to year
jpx = jp["Year"]
jpy = jp["value in million tons of oil equivalent"]

plt.figure(figsize=(20, 15))

plt.plot(cnx,cny,label = "China")
plt.plot(usx,usy,label = "The United States")
plt.plot(trfx,trfy,label = "The Russian Federation")
plt.plot(indx,indy,label = "India")
plt.plot(jpx,jpy,label = "Japan")
plt.ylabel("Total energy consumption during 2007-2016",size = 18)
plt.legend()

#As we can see in the graph.China and India are the two countries that has a increasing energy consumption during 2007-2016.
#This coherent the assumption from the last two histogram that energy consumption of most countries don't fluctuate that much.
```

# Consumption level distribution

```python
df3 = df.drop(["YoY in 2016"],axis = 1)
df3 = df3[df3["Country"] != "Global total"]
df3["value in million tons of oil equivalent"].plot(kind="hist",figsize = (14,8))
plt.xlabel("Yearly energy consumption")
plt.show()
#Each column represent the range of each country's energy consumption of each year during 2006-2017. 
#By using histogram to plot out their frequency, we can see that for most countries, their yearly energy consumption is around 100-500 million tons 
```

# Top countries comparison

```python
df5 = df.drop(["YoY in 2016"],axis = 1)

x1=df5[(df5['Country'] == 'China')]["value in million tons of oil equivalent"]

x2=df5[(df5['Country'] == 'United States')]["value in million tons of oil equivalent"]

x3=df5[(df5['Country'] == 'The Russian Federation')]["value in million tons of oil equivalent"]

x4=df5[(df5['Country'] == 'India')]["value in million tons of oil equivalent"]

x5=df5[(df5['Country'] == 'Japan')]["value in million tons of oil equivalent"]

plt.figure(figsize=(20, 8))

plt.subplot(1, 5, 1)
x1.plot(kind='box',color='red',title  = 'China')

plt.subplot(1, 5, 2)
x2.plot(kind='box',color='blue',title  = 'United States')

plt.subplot(1, 5, 3)
x3.plot(kind='box',color='green',title  = 'The Russian Federation')

plt.subplot(1, 5, 4)
x4.plot(kind='box',color='purple',title  = 'India')

plt.subplot(1, 5, 5)
x4.plot(kind='box',color='orange',title  = 'Japan')

plt.show()

#I picked some big countries and analyze their data distribution by the boxplot.
#The box plot shows that during 2007-2016, the energy consumption level of USA and Russia are relatively stable.However,China, India, and Japan 
#shows that there are certain changes from 2007 to 2017, and we know that the energy consumption level of Japan decreased,but energy consumption
#level of India and China had increased.
```
